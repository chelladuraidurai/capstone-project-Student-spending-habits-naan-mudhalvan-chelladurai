```python
categorical_columns = [col for col in df.columns if df[col].dtype=='0']
numerical_columns = [col for col in df.columns if df[col].dtype!='0']
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize = (12, 7))
sns.histplot(data = df,
            x = 'age',
            kde = True,
            bins = round(np.sqrt(len(df))))
plt.title("Student's age")
plt.show()
```
